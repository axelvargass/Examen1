```python
#Lectura de archivo tipo csv
lastmile = pd.read_csv('lastmile_delivery_operations_mit_lift_lab.csv')
lastmile
```

```python
#Filtro de columnas para cualitativas
cualitativas = lastmile.iloc[:,[0,1,2,3,4,8,9,10,11,13,14,15,17,18,19,20,23,24,26,27]]
```

```python
#Filtro de columnas para cualitativas
cuantitativas = lastmile.iloc[:,[5,6,7,12,16,21,28]]
```

```python
#Corroboramos valores nulos
valores_nulos = cualitativas.isnull().sum()
valores_nulos 
```

```python
#Limpiamos Valores nulos de cualitativos
filtrocualitativos = cualitativas.fillna(method='ffill')
```

```python
#Corroboramos valores nulos de Cualitativos para limpiarlos
valores_nulos = filtrocualitativos.isnull().sum()
valores_nulos
```

```python
#Corroboramos valores nulos MAL!!
valores_nulos = filtrocualitativos.isnull().sum()
valores_nulos
```

```python
#Limpiamos los valores nulos de Cuantitativas
filtrocuantitativas = cuantitativas.fillna(round(cuantitativas.mean(),1))
```

```python
#Corroboramos valores nulos de Cualitativos para limpiarlos
valores_nulos_cuantitativos = filtrocuantitativas.isnull().sum()
valores_nulos
```

```python
#Metodo de IQR
y = filtrocuantitativas
percentile25 = y.quantile(0.25) #Q1
percentile75 = y.quantile(0.75) #Q3
iqr = percentile75 - percentile25

Limite_Superior_iqr = percentile75 + 1.5*iqr
Limite_Inferior_iqr = percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Obtenemos datos limpios
data_clean_iqr = filtrocuantitativas[(y <= Limite_Superior_iqr)&(y >= Limite_Inferior_iqr)]
data_clean_iqr
```

```python
#Unimos ambos dataframes ya que estan limpios
datos_limpios = pd.concat([filtrocuantitativas,filtrocualitativos],axis=1)
datos_limpios
```

```python
#Convertir dataframe a CSV
datos_limpios.to_csv("Lastmile_Delivery.csv")
#Descargar archivo filtrado en csv
from google.colab import files


files.download("Lastmile_Delivery.csv")
```
